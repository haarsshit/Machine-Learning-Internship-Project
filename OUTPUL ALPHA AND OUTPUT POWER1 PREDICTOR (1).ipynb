```python
from sklearn.model_selection import train_test_split
```

```python
x = D.iloc[:,:-3]
```

```python
y = D.iloc[:,-2:]
```

```python
x.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
mult_reg = MultiOutputRegressor(GradientBoostingRegressor(random_state=0)).fit(X_train, y_train)
```

```python
y_pred = mult_reg.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error
```

```python
train_pred = mult_reg.predict(X_train)
```

```python
train_score = mean_squared_error(y_train,train_pred)
```

```python
train_score
```

```python
test_score = mean_squared_error(y_test,y_pred)
```
